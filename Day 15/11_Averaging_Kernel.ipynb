## Averaging Kernel

```python
import cv2
import numpy as np
```

```python
img = cv2.imread(r'/home/pgdai/Desktop/March24-DAI-Workspace/CV-June24/images/lenna.png')

#Define a 5x5 averaging Kernel
kernel = np.ones((5,5),np.float32)/25

# Apply the kernel to smooth the image
smoothed_image = cv2.filter2D(img, -1, kernel)

# Show the original and smoothed image
cv2.imshow('Original Image', img)
cv2.imshow("Smoothed Image:", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
new_image = np.hstack([img, smoothed_image])
```

```python
cv2.imshow("Both Image:", new_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Gaussian Kernel

```python
kernel_size = 7   # Adjust as needed
sigma = 0.35 * ((kernel_size-1)*0.5-1)+0.9     # Gaussian standard deviation
kernel = cv2.getGaussianKernel(kernel_size,sigma)
```

### Median Filter 

```python
import matplotlib.pyplot as plt
```

```python
gray = cv2.imread('../Day14/images/lenna.png', 0)
```

```python
gray.shape
```

#### Blurring Effect

```python
kernel_size = 5
median = cv2.medianBlur(gray, kernel_size)
plt.figure(figsize=(10, 6))
plt.subplot(121), plt.imshow(gray, cmap='gray'), plt.title("Original Image")
plt.subplot(122), plt.imshow(median, cmap='gray'), plt.title("Median Filtered Image")
plt.show()
```
