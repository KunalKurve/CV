```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r'D:\NLP\Day14\images\lenna.png',0)
```

```python
plt.imshow(img, cmap='Reds');
```

## Sequential Colormaps

```python
plt.figure(figsize=(16, 8))
plt.subplot(2, 4, 1)
plt.title('Reds')
plt.imshow(img, cmap='Reds')
plt.subplot(2, 4, 2)
plt.title('Blues')
plt.imshow(img, cmap='Blues')
plt.subplot(2, 4, 3)
plt.title('Greens')
plt.imshow(img, cmap='Greens')
plt.subplot(2, 4, 4)
plt.title('Oranges')
plt.imshow(img, cmap='Oranges')
plt.subplot(2, 4, 5)
plt.title('Purples')
plt.imshow(img, cmap='Purples')
plt.subplot(2, 4, 6)
plt.title('OrRd')
plt.imshow(img, cmap='OrRd')
plt.subplot(2, 4, 7)
plt.title('YlGn')
plt.imshow(img, cmap='YlGn')
plt.subplot(2, 4, 8)
plt.title('PuRd')
plt.imshow(img, cmap='PuRd')
```

## Diversing Colormaps

```python
plt.figure(figsize=(16, 8))
plt.subplot(2, 4, 1)
plt.title('PiYG')
plt.imshow(img, cmap='PiYG')
plt.subplot(2, 4, 2)
plt.title('coolwarm')
plt.imshow(img, cmap='coolwarm')
plt.subplot(2, 4, 3)
plt.title('bwr')
plt.imshow(img, cmap='bwr')
plt.subplot(2, 4, 4)
plt.title('seismic')
plt.imshow(img, cmap='seismic')
plt.subplot(2, 4, 5)
plt.title('PuOr')
plt.imshow(img, cmap='PuOr')
plt.subplot(2, 4, 6)
plt.title('RdGy')
plt.imshow(img, cmap='RdGy')
plt.subplot(2, 4, 7)
plt.title('Spectral')
plt.imshow(img, cmap='Spectral')
plt.subplot(2, 4, 8)
plt.title('RdBu')
plt.imshow(img, cmap='RdBu')
plt.savefig('diversing.png')
```

## Cyclic Colormaps

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 3, 1)
plt.title('twilight')
plt.imshow(img, cmap='twilight')
plt.subplot(1, 3, 2)
plt.title('twilight_shifted')
plt.imshow(img, cmap='twilight_shifted')
plt.subplot(1, 3, 3)
plt.title('hsv')
plt.imshow(img, cmap='tab10')
```

## Quantitative Colormaps

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
plt.title('tab10')
plt.imshow(img, cmap='tab10')
plt.subplot(1, 2, 2)
plt.title('Set2')
plt.imshow(img, cmap='Set2')
```
