```python
# Q4. Download the video nature.mp4
# and Load the video
video_path = "/Users/kunalkurve/CDAC-PGDAI/All Modules/NLP/nature.mp4"
cap = cv2.VideoCapture(video_path)
```

```python
# 1. Find height, width, total frames and fps of the video
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = cap.get(cv2.CAP_PROP_FPS)

print("Video Resolution:", height,width)
print("Total Frame:", total_frames)
print('FPS', fps)
```

```python
#### 2. Play the video in frames

cap = cv2.VideoCapture(video_path)
while True:
    ret, frame = cap.read()
    if not ret:
        break
    
    #Process thr frame e.g -> apply filters, detect objects
    
    cv2.imshow('Video', frame)
    if cv2.waitKey(5) == ord('q'):  #ord -> ascii value of 'q'
        break
cv2.destroyAllWindows()
```

```python
# 3. Save the image at the 50th second (assuming 30 fps)

frame_to_save = int(50 * fps)
cap = cv2.VideoCapture(video_path)

cap.set(cv2.CAP_PROP_POS_FRAMES, frame_to_save)
ret, frame = cap.read()

if ret:
    cv2.imwrite('frame_at_50thsecond.jpg', frame)

cap.release()
```
