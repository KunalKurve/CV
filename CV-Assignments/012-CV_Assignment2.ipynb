```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('sachin.jpg')
```

1. Apply following transformations on the image.
1. Translate the image towards right by 400 and up by 150 pixels.
2. Rotate the image 60 degrees anti clockwise.
3. Resize the image to 75% of original dimensions.
4. Flip the image horizontally, diagonally and vertically. Show all four
images in the row.
5. Crop the image at, 500,300 to 560, 400 and save in another file

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 1. Translate the image towards right by 400 and up by 150 pixels.

# Translation Matrix
M = np.float32([[1, 0, 400], [0, 1, 150]])

# Apply the translation to the image
translated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Display the translated image
plt.imshow(cv2.cvtColor(translated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 2. Rotate the image 60 degrees anti clockwise.

# Rotation Matrix
M = cv2.getRotationMatrix2D((img.shape[1] // 2, img.shape[0] // 2), -60, 1.0)
# Apply the rotation to the image
rotated_img = cv2.warpAffine(img, M, (img.shape[1], img.shape[0]))

# Display the rotated image
plt.imshow(cv2.cvtColor(rotated_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 3. Resize the image to 75% of original dimensions.

resized_img = cv2.resize(img, None, fx=0.75, fy=0.75)

# Display the resized image
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
# 4.Flip the image horizontally, diagonally and vertically. Show all four images in the row.

flipped_horizontally = cv2.flip(img, 1)
flipped_vertically = cv2.flip(img, 0)

# For diagonal flip, rotate by 90 then flip horizontally
rotated_90 = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
flipped_diagonally = cv2.flip(rotated_90, 1)
```

```python
# Display images using matplotlib
fig, axes = plt.subplots(1, 3, figsize=(20, 5))
axes[0].imshow(cv2.cvtColor(flipped_horizontally, cv2.COLOR_BGR2RGB))
axes[0].set_title('Flipped Horizontally')
axes[1].imshow(cv2.cvtColor(flipped_vertically, cv2.COLOR_BGR2RGB))
axes[1].set_title('Flipped Vertically')
axes[2].imshow(cv2.cvtColor(flipped_diagonally, cv2.COLOR_BGR2RGB))
axes[2].set_title('Flipped Diagonally')
plt.show()
```

```python
# 5. Crop the image at, 500,300 to 560, 400 and save in another file

cropped_img = img[300:400, 500:560]
plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.axis('off')
cv2.imwrite('cropped_sachin.jpg', cropped_img)
```
